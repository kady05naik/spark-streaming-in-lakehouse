```python
data_path='/Volumes/cat_spark_streaming/wordcnt/wordcnt/'
table_path='cat_spark_streaming.wordcnt'
```

```python
lines=spark.read \
    .format('text') \
    .option('multiLine', True) \
    .option('lineSep', '.') \
    .load(data_path)
```

## 2. Processing Data

```python
from pyspark.sql.functions import explode,split,trim,lower,col,count
```

Trim the spaces from both ends for the specified string column

```python
raw_words=lines.select(explode(split(col('value')," ")).alias('word'))

quality_words=raw_words.select(trim(col('word')).alias('word') )

quality_words=quality_words.select((split(col('word'),',')[0]).alias('word'))

quality_words=quality_words.select(col('word')) \
                        .where("word is not null") \
                        .where("word rlike '[a-z]'") \
                
```

```python
quality_words=quality_words.select(col('word')) \
                        .where("word is not null") \
                        .where("word rlike '[a-z]'") 
```

```python
quality_words.display()
```

```python
WordCounts=quality_words \
        .groupBy('word') \
        .agg(count(col('word')).alias('wordCount'))
```

```python
WordCounts.display()
```

```python
WordCounts.display()
```

```python
WordCounts.write \
            .mode('overwrite') \
            .format('delta') \
            .saveAsTable(f'{table_path}.WordCountBatch') 
```
